```python
import matplotlib
from matplotlib import pyplot
%matplotlib inline
```

```python
from utils import save_dir
import cPickle
```

```python
experiment_name = 'more_capacity'
```

```python
import os
from blocks.serialization import load

with open(os.path.join(
        save_dir, 'config',
        experiment_name + '.pkl')) as f:
    saved_args = cPickle.load(f)

with open(os.path.join(
        save_dir, "pkl",
        "best_" + experiment_name + ".tar"), 'rb') as src:
    log = load(src, name='log')
```

```python
channels = [['train_nll', 'valid_nll']]
all_channels = list(set([x for small in channels for x in small]))
num_plots = len(channels)

from pandas import DataFrame
df = DataFrame.from_dict(log, orient='index')
df = df[all_channels].astype(float)
df = df.interpolate('index')

if num_plots > 1:
    fig, axarr = pyplot.subplots(num_plots, sharex=True)
    for i, channel in enumerate(channels):
        df[channel].plot(ax=axarr[i])
else:
    df[channels[0]].plot()
```
